```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
import preprocessingFunction 
```

```python
raw_data = pd.read_csv('train.csv', keep_default_na = False)
raw_data.head()
```

```python
data = raw_data.copy()
data.head()
```

```python
data2 = preprocessingFunction.preprocessinghandle(data)
```

```python
data2.head()
```

# Calculate the Pearson correlation of the features

```python
data3 = data2.copy()
data3 = data3.drop(['Id'], axis = 1)
sale_price = np.array(data3['SalePrice'])
data3 = data3.drop(['SalePrice'], axis = 1)
data3_columns = data3.columns.values
pearson_correlation_array = []
for i in data3_columns:
    pearson_correlation_array.append(abs(pearsonr(sale_price, data3[i])[0]))
    #print(abs(pearsonr(sale_price, data3[i])[0]), i)
pearson_correlation_array = np.array(pearson_correlation_array)
```

# Creating the data frame of the correlation

```python
correlation_dataframe = pd.DataFrame({'Feature': data3_columns, 'Pearson Correlation': pearson_correlation_array})
correlation_dataframe.sort_values(by=['Pearson Correlation'])
```

```python
features_to_drop = []
for i in range(len(correlation_dataframe['Pearson Correlation'])):
    if correlation_dataframe['Pearson Correlation'][i] < 0.3:
        features_to_drop.append(correlation_dataframe['Feature'][i])
```
