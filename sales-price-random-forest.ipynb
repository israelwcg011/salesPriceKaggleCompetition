# Important libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import preprocessingFunction as app2
import helpers as hp
```

# Reading the data

```python
#dataTrain = pd.read_csv('train.csv', keep_default_na = False)
#dataTest = pd.read_csv('test.csv', keep_default_na = False)
dataTrain = pd.read_csv('train.csv')
dataTrain.fillna('NA')
dataTest = pd.read_csv('test.csv')
dataTest.fillna('NA')
```

## Preprocessing

```python
columnsToDrop = ['Id', 'Alley', 'PoolQC', 'Fence', 'MiscFeature', 'MSZoning', 'Condition2', 'HouseStyle', 'RoofMatl', 'Exterior1st', 'Heating', 'Electrical', 'KitchenQual', 'Functional', 'GarageQual', 'SaleType', 'Utilities', 'Exterior2nd']
#columnsToDrop = ['Id', 'Alley', 'PoolQC', 'Fence', 'MiscFeature']
```

```python
#train data
dfTrain = dataTrain.copy()
dfTrain = dfTrain.drop(columnsToDrop, axis = 1)

#test data
dfTest = dataTest.copy()
dfTest = dfTest.drop(columnsToDrop, axis = 1)
```

```python
#train data
dfTrain2 = app2.preprocessinghandle(dfTrain)
dfTrain2 = dfTrain2.sample(frac=1).reset_index(drop=True)

#test data
dfTest2 = app2.preprocessinghandle(dfTest)
dfTest2 = dfTest2.sample(frac=1).reset_index(drop=True)
```

```python
dfTrain2.head()
```

```python
dfTest2.head()
```

```python
# labels
labels = np.array(dfTrain2['SalePrice'])
# Remove the labels from the dataFrame
dfTrain2 = dfTrain2.drop('SalePrice', axis = 1)
```

```python
testColumnsArray = dfTest2.columns.values
trainColumnsArray = dfTrain2.columns.values
tempColumns = []

for i in trainColumnsArray:
    if i not in testColumnsArray:
        tempColumns.append(i)
```

```python
dfTrain2 = dfTrain2.drop(tempColumns, axis = 1)
dfTrain2 = dfTrain2.reindex(columns= testColumnsArray)
```

# Features and Targets and Convert Data to Arrays

```python
dfTrain3 = dfTrain2.copy()
dfTest3 = dfTest2.copy()
```

```python
# labels
#  labels = np.array(dfTrain3['SalePrice'])
# Remove the labels from the dataFrame
#  dfTrain3 = dfTrain3.drop('SalePrice', axis = 1)
# Saving feature names for later use
feature_list = list(dfTrain3.columns)
# Convert to numpy array
dfTrain3 = np.array(dfTrain3)
features = dfTrain3

features_data_test = np.array(dfTest3)
```

# Split the data into training and testing

```python
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.15, random_state = None)
```

# Train model

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 2000, random_state = None)
# Train the model on training data
rf.fit(train_features, train_labels);
```

# Test model

```python
# Use the forest's predict method on the test data
predictions = rf.predict(test_features)
# Calculate the absolute errors
errors = abs(predictions - test_labels)
```

# Accuracy

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / test_labels)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
